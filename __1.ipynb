Данное распределение не является нормальным. Оно бимодальное (имеет два ярко выраженных пика) или мультимодальное. Данные не распределены равномерно по обе стороны от центрального значения, и кривая плотности не похожа на ровный колокол, так как имеет два явных "горба" (пика), а в центре провал.

```python
# ниже код для оптимизации вывода всех числовых переменных
num_cols = df.columns.intersection(['funding_per_student_usd', 'dropout_rate_percent', 'percent_low_income'])
fig, axes = plt.subplots(1, len(num_cols), figsize=(6 * len(num_cols), 4))

for ax, col in zip(axes, num_cols):
    sns.histplot(df[col], kde=True, ax=ax)
    ax.set_title(f'Распределение {col}')
    ax.set_xlabel(col)
    ax.set_ylabel('Частота')

plt.tight_layout()
plt.show()
```

Все три графика показывают распределения, отличные от нормального (распределения несимметричны и не имеют единого центрального пика, похожего на колокол).

1.Данное распределение является бимодальным (имеет два явных пика). Наибольшая частота наблюдается в двух диапазонах. Первый крупный пик приходится на период от 7 500 до 12 500 USD. Второй крупный пик приходится на период от 20 000 до 25 000 USD.

2.Распределение отличается от нормального. Оно слабо асимметрично влево (левый хвост более выражен), с более высокой концентрацией значений в правом диапазоне.Наибольшая частота наблюдается в период от 40 до 90 процентов.

3.Распределение отлично от нормального. Оно мультимодальное (имеет несколько выраженных пиков).Наблюдаются три основных пика частоты в диапазонах от 2 до 4%, от 6 до 8%, и от 10 до 12% отсева.

2. **Столбчатая диаграмма для категориальной переменной `Class`:**

**Боксплот для числовой переменной:**

```python
num_cols = df.select_dtypes(include='float').columns
fig, axes = plt.subplots(1, len(num_cols), figsize=(6 * len(num_cols), 4))

for ax, col in zip(axes, num_cols):
    sns.boxplot(df[col], ax=ax)
    ax.set_title(f'Распределение {col}')
    ax.set_xlabel(col)
    ax.set_ylabel('Частота')

plt.tight_layout()
plt.show()
```
